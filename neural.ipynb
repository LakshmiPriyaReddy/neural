```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error
```

```python
# Load the data
data = pd.read_csv('datacleaned_running_data.csv')
```

```python
# Prepare the features and the target
features = data.drop(['Activity ID', 'Activity Name', 'Activity Type', 'Commute', 'Filename', 'Athlete Weight','Activity Date',
                     'Average Grade', 'Commute.1'], axis=1)
target = data['Elapsed Time']
```

```python
features
```

```python
# Split the data into training and testing sets
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
features_train = scaler.fit_transform(features_train)
features_test = scaler.transform(features_test)
```

```python
# Create the model
model = Sequential()
model.add(Dense(64, input_dim=features_train.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Train the model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5)
model.fit(features_train, target_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
# Make predictions
predictions = model.predict(features_test)
```

```python
# Calculate mean squared error
mse = mean_squared_error(target_test, predictions)
print('Mean Squared Error:', mse)
```
